## Determinant Calculator - Using Recursion

A determinant is a value associated to a square array of numbers, that square array being called a square matrix.

### Determinant of a 2 by 2 matrix

#### Example:

Implementing example

```python
def det2by2(m):
    return  m[0][0] * m[1][1] - m[1][0] * m[0][1]
```

```python
mat1 = [[2, 4],
       [-3, 1]]
```

```python
print(det2by2(mat1))
```

implementing function minor

```python
def getminor(m, i, j):
    return [row[: j] + row[j+1:] for row in (m[: i] + m[i+1:])]
```

### Determinant Definition

# Example:

Implementing determinant in a using loop

```python
def det(m,idx):
    #idx is index of square matrix
    result = 0
    for cofrow in range (0, idx):
        result = result + ((-1)**(cofrow)* m[0][cofrow]*det2by2(getminor(m, 0, cofrow)))
    return result
```

Implementing determinant in recursion

```python
def detrec(m,idx):
    #idx is index of square matrix
    result = 0
    
    #return base case when idx is 2 (Square matrix)
    if (idx == 2):
        return det2by2(m)
    else:
        for cofrow in range (0, idx):
            sign = (-1)**(cofrow)      
            result = result + sign*m[0][cofrow]*detrec(getminor(m, 0, cofrow),idx -1)
        return result
```

Implementing determinant in recursion with only one input

```python
mat3 = [[1, 2, -1],
       [3, 0, 1],
       [4, 2, 1]]
```

```python
det(mat3,3)
```

```python
detrec(mat3,3)
```

```python
mat4 = [[1, 0, 2, -1],
       [3, 0, 0, 5],
       [2, 1, 4, -3],
       [1, 0, 5, 0]]
```

```python
detrec(mat4,4)
```
